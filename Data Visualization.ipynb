Q1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

```python
import plotly.graph_objects as go
import numpy as np

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
fig = go.Figure()
fig.add_trace(go.Scatter3d(x=df['X'], y=df['Y'], z=df['Z'], mode='markers'))
fig.update_layout(scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))
fig.show()
```

Q3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
days.

```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
heatmap_data = df.pivot_table(index='Month', columns='Day', values='Sales', aggfunc='mean', fill_value=0)
fig = go.Figure(data=go.Heatmap(z=heatmap_data.values, x=heatmap_data.columns, y=heatmap_data.index))
fig.update_layout(title='Variation in Sales Across Months and Days')
fig.show()
```

Q4. Using the given x and y data, generate a 3D surface plot to visualize the function

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
fig = go.Figure(data=[go.Surface(z=df['Z'].values.reshape(100, 100))])
fig.update_layout(scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))
fig.show()
```

Q1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

p.line(x, y, legend_label="sine", line_width=2)

show(p)
```

Q5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)

y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)

```python
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource, ColorBar, LinearColorMapper
from bokeh.palettes import Viridis256
import numpy as np
import pandas as pd

# Sample Data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare data
data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten(),
    'width': np.ones_like(xx.flatten()) * (x[1] - x[0]),  # Width of each cell
    'height': np.ones_like(yy.flatten()) * (y[1] - y[0])  # Height of each cell
}
# Calculate right positions
data['right'] = data['x'] + data['width']
data['top'] = data['y'] + data['height']

source = ColumnDataSource(data)

# Create a Bokeh figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', x_range=(0, 1), y_range=(0, 1))

# Define color mapper
min_value = np.min(data_heatmap)
max_value = np.max(data_heatmap)
color_mapper = LinearColorMapper(palette=Viridis256, low=min_value, high=max_value)

# Add quad glyph for heatmap
p.quad(
    top='top',
    bottom='y',
    left='x',
    right='right',
    source=source,
    fill_color={'field': 'value', 'transform': color_mapper},
    line_color=None
)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Output to file and show the plot
output_file('heatmap.html')
show(p)
```
